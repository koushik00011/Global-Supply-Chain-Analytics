```python
df['customer_zipcode']=df['customer_zipcode'].fillna(df['customer_zipcode'].mean())
```

```python
df['customer_lname']=df['customer_lname'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
df['customer_zipcode'].dtype
```

```python
#Delivery Gap Calculation
#Delay Gap 
df['delay_gap']=df['days_for_shipping_(real)']-df['days_for_shipment_(scheduled)']
```

```python
df.head()
```

```python
def classification(gap):
    if gap<0:
        return 'early'
    elif gap==0:
        return 'on-time'
    else:
        return 'delayed'
df['delivery_category']=df['delay_gap'].apply(classification)
```

```python
df['delivery_category'].value_counts()
```

```python
df.head()
```

```python
df=df.rename(columns={'days_for_shipping_(real)': 'actual_shipping_days',
    'days_for_shipment_(scheduled)': 'scheduled_shipping_days'})
```
